```python
import pandas as pd
import sys
```

```python
df = pd.read_csv("../data/stack-overflow-developer-survey-2025/survey_results_public.csv")
print("Shape:", df.shape)
print("\n Columnas:", df.columns.tolist())
print("\n Nulos por columna (%):")
print((df.isnull().sum() / len(df) * 100).sort_values(ascending=False).head(30))
print("\n Primeras filas:")
print(df.head(2))
```

```python
# Variables claves
variables_claves = ["AIAgents",
                    "AIAgentChange",
                    "AIThreat", 
                    "NewRole",
                    "AISelect",
                    "AISent",
                    "JobSat",
                    "Country",
                    "DevType"]

with open("variables_clave.txt", "w", encoding="utf-8") as f:
    for c in variables_claves:
        f.write(f"\n{'-'*50}\n")
        f.write(f"{c} - nulos: {df[c].isnull().sum()} ({df[c].isnull().mean()*100:.1f}%)\n")
        f.write(str(df[c].value_counts().head(15)) + "\n")

print("Variables guardadas en txt")
```
